```python
import os
import novas
import numpy as np
from sqlite3 import OperationalError
from sifter import precalc as p
```

```python
fileName = '../dev_data/tot.itf'
```

```python
with open(fileName) as f:
  lineListFull = f.readlines()
```

```python
lineList = lineListFull[:500]
```

```python
try:
    T = p.Tracklets(lineList)
except OperationalError:
    T = p.Tracklets()
    all_obs = T.parse_all_observations(lineList)
except KeyError:
    print("*" * 80 + "\n" +
          "*" * 30 + "       ERROR!       " + "*" * 30 + "\n" +
          "*" * 80 + "\n" +
          "*" * 20 + " That's not a valid Observatory Code!!! " + "*" * 20 + "\n" +
          "*" * 80 + "\n")
    raise
```
